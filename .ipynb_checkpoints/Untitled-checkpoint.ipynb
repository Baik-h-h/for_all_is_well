```python
import requests
import xml.etree.ElementTree as ET
import time

error_retry_interval = 60
retry_interval = 60 * 3

time_lst=[]
for i in range(4,24):
    for j in range(0,60,10):
        i = str(i)
        j = str(j)
        if len(i) == 1:
            i = '0'+i
        if len(j) == 1:
            j = '0'+j
        time_lst.append(f'2024-05-02 {i}:{j}')
        
#print(time_lst)
k=0
while True:
    print('')
    # API 엔드포인트
    url = "http://openapi.seoul.go.kr:8088/4465666c63626c613130394f75596657/xml/citydata/1/5/%EB%AA%85%EB%8F%99%20%EA%B4%80%EA%B4%91%ED%8A%B9%EA%B5%AC"
    
    try:
        # API에 요청을 보내고 응답 받기
        response = requests.get(url)
        response.raise_for_status()  # 응답이 성공적이지 않으면 예외 발생
        
        # XML 데이터 파싱
        xml_data = response.text
        root = ET.fromstring(xml_data)
        
        # XML 데이터에서 <PPLTN_TIME> 요소 찾기
        ppltn_time = root.find('.//PPLTN_TIME')
        
        # 찾은 요소의 텍스트 출력
        if ppltn_time is not None:
            print('접속 시간:',time.strftime('%Y-%m-%d %H:%M:%S'))
            print("인구 조사 시간:", ppltn_time.text)
            if ppltn_time.text == time_lst[k]:
                file_name = time_lst[k].replace(':', '_') + ".xml"  # 콜론을 언더스코어로 대체하여 파일 이름 생성
                with open(file_name, "w", encoding="utf-8") as file:
                    file.write(xml_data)
                print(f'성공!! 다음시간 {time_lst[k+1]}로 넘어갑니다.')
                k += 1
                #print(k)
                time.sleep(retry_interval)
                
            else:
                print(f'아직 {time_lst[k]}의 정보가 업데이트 되지 않음')
                tmp_num=0
                for tm in time_lst[k+1:-1]:
                    tmp_num+=1
                    if tm == ppltn_time.text:
                        print(f'{time_lst[k]}정보가 없어서 {tm}으로 넘어갑니다.')
                        file_name = tm.replace(':', '_') + ".xml"  # 콜론을 언더스코어로 대체하여 파일 이름 생성
                        with open(file_name, "w", encoding="utf-8") as file:
                            file.write(xml_data)
                        k+=(tmp_num+1)
                        #print(k)
                time.sleep(retry_interval)
            
        else:
            print("<PPLTN_TIME> 요소를 찾을 수 없습니다.")
            time.sleep(error_retry_interval)
            break  # 루프 종료
    except requests.exceptions.RequestException as e:
        print("API 요청에 실패하였습니다:", e)
        print("1분 후에 다시 시도합니다.")
        time.sleep(error_retry_interval)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import warnings

warnings.filterwarnings('ignore')

# 새로운 데이터프레임을 생성합니다.
columns = ['time', 'AREA_PPLTN_MIN', 'AREA_PPLTN_MAX', 'MALE_PPLTN_RATE', 'FEMALE_PPLTN_RATE',
           'PPLTN_RATE_0', 'PPLTN_RATE_10', 'PPLTN_RATE_20', 'PPLTN_RATE_30', 'PPLTN_RATE_40',
           'PPLTN_RATE_50', 'PPLTN_RATE_60', 'PPLTN_RATE_70', 'RESNT_PPLTN_RATE', 'NON_RESNT_PPLTN_RATE']
df = pd.DataFrame(columns=columns)

# 파일 이름 형식에 따라 모든 XML 파일을 불러와 데이터프레임에 추가합니다.
for hour in range(6, 24):
    for minute in range(0, 60, 10):
        hour_str = str(hour).zfill(2)
        minute_str = str(minute).zfill(2)
        file_name = f'2024-05-02 {hour_str}_{minute_str}.xml'
        
        try:
            # XML 파일을 파싱합니다.
            tree = ET.parse(file_name)
            root = tree.getroot()
            
            # XML 데이터에서 필요한 부분을 추출합니다.
            data = [hour_str + ':' + minute_str]
            for column in columns[1:]:
                value = root.find(f".//{column}").text
                data.append(value)
            
            # 데이터를 데이터프레임에 추가합니다.
            df = df.append(pd.Series(data, index=columns), ignore_index=True)
        except FileNotFoundError:
            print(f"{file_name}을(를) 찾을 수 없습니다.")

            
df['AREA_PPLTN_AVG'] = (df['AREA_PPLTN_MIN'].astype(float) + df['AREA_PPLTN_MAX'].astype(float)) / 2


print(df)
```

```python
import matplotlib.pyplot as plt

# 그래프를 그릴 데이터 선택
data = df[['time', 'AREA_PPLTN_AVG', 'MALE_PPLTN_RATE', 'FEMALE_PPLTN_RATE']]

# 그래프 그리기
plt.figure(figsize=(12, 8))

# 'AREA_PPLTN_AVG' 그래프
plt.plot(data['time'], data['AREA_PPLTN_AVG'], label='AREA_PPLTN_AVG')

# 'MALE_PPLTN_RATE' 그래프
plt.plot(data['time'], data['MALE_PPLTN_RATE'], label='MALE_PPLTN_RATE')

# 'FEMALE_PPLTN_RATE' 그래프
plt.plot(data['time'], data['FEMALE_PPLTN_RATE'], label='FEMALE_PPLTN_RATE')

# 그래프 제목과 레이블 설정
plt.title('Population Data by Time')
plt.xlabel('Time')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

```

```python
# 새로운 DataFrame df_sum 생성
df_sum = pd.DataFrame(columns=df.columns)

# df의 행을 3개씩 묶어서 평균을 계산하고 df_sum에 추가
for i in range(0, len(df), 3):
    # 행의 시작과 끝 인덱스 계산
    start_index = i
    end_index = min(i + 3, len(df))
    
    # 시간열을 제외한 나머지 열의 평균 계산
    mean_values = df.iloc[start_index:end_index, 1:].astype(float).mean(axis=0)
    
    # 첫 번째 행의 'time' 열 값 가져오기
    time_value = df.iloc[start_index]['time']
    
    # 새로운 행 추가
    df_sum = df_sum.append(pd.Series([time_value] + mean_values.tolist(), index=df.columns), ignore_index=True)

# 결과 확인
print(df_sum)
```

```python
# 각 시간대별로 가장 많은 비율을 차지하는 집단을 저장할 딕셔너리
most_common_groups = {}
con_1 = ['MALE_PPLTN_RATE', 'FEMALE_PPLTN_RATE']
con_2 = ['RESNT_PPLTN_RATE', 'NON_RESNT_PPLTN_RATE']
con_3 = ['PPLTN_RATE_0', 'PPLTN_RATE_10', 'PPLTN_RATE_20', 'PPLTN_RATE_30', 'PPLTN_RATE_40','PPLTN_RATE_50', 'PPLTN_RATE_60', 'PPLTN_RATE_70']
# 시간대별로 인구 데이터 분석
for index, row in df_sum.iterrows():
    time = row['time']
    max_ratio = 0
    most_common_group = [0,0,0]
    second_max = 0
    third_max = 0
    
    
    # 각 집단별 비율을 확인하고 가장 높은 비율을 가지는 집단을 결정
    for i in range(len(con_1)):
        for j in range(len(con_2)):
            for k in range(len(con_3)):
                ratio = float(row[con_1[i]])*float(row[con_2[j]])*float(row[con_3[k]])
                if ratio > max_ratio:
                    max_ratio = ratio
                    most_common_group[0] = 100*i+10*j+k
                elif ratio > second_max:
                    second_max = ratio
                    most_common_group[1] = 100*i+10*j+k
                elif ratio > third_max:
                    third_max = ratio
                    most_common_group[2] = 100*i+10*j+k
    
    # 결과를 딕셔너리에 저장
    most_common_groups[time] = most_common_group

    
#print(most_common_groups)
# 결과 출력
for time, groups in most_common_groups.items():
    #print(group)
    group_output=['','','']
    tmp_num=0
    for group in groups:
        if group//100 == 1:
            group_output[tmp_num] = 'Female_'
            group = group%100
        else:
            group_output[tmp_num] = 'Male_'

        if group // 10 == 1:
            group_output[tmp_num]+='Resident_'
            group = group%10
        else:
            group_output[tmp_num] += 'Nonresident'

        if group==0:
            group_output[tmp_num]+="0's"
        elif group==1:
            group_output[tmp_num]+="10's"
        elif group==2:
            group_output[tmp_num]+="20's"
        elif group==3:
            group_output[tmp_num]+="30's"
        elif group==4:
            group_output[tmp_num]+="40's"
        elif group==5:
            group_output[tmp_num]+="50's"
        elif group==6:
            group_output[tmp_num]+="60's"
        elif group==7:
            group_output[tmp_num]+="70's"
        tmp_num+=1
    output_lst = ['most','second','third']
    for output in range(len(output_lst)):
        print(f"At {time}, the {output_lst[output]} common group is {group_output[output]}.")
    print()
```

```python
import pandas as pd

# 데이터프레임 생성
data_20 = pd.read_csv('data_20.csv')
data_30 = pd.read_csv('data_30.csv')
data_40 = pd.read_csv('data_40.csv')
data_all = pd.read_csv('data_all.csv')

# 데이터프레임 확인
print(data_20)
print(data_30)
print(data_40)
print(data_all)
```

```python
# 'Date' 열이 '2022-12-25'인 행 선택
selected_row_20 = data_20[data_20['Date'] == '2022-12-25']
selected_row_30 = data_30[data_30['Date'] == '2022-12-25']
selected_row_40 = data_40[data_40['Date'] == '2022-12-25']
selected_row_all = data_all[data_all['Date'] == '2022-12-25']

# 'Four_cuts', 'Photoism', 'Haru_firm' 열의 값을 확인
values_20 = selected_row_20[['Four_cuts', 'Photoism', 'Haru_firm']]
values_30 = selected_row_30[['Four_cuts', 'Photoism', 'Haru_firm']]
values_40 = selected_row_40[['Four_cuts', 'Photoism', 'Haru_firm']]
values_all = selected_row_all[['Four_cuts', 'Photoism', 'Haru_firm']]

rows=['Four_cuts', 'Photoism', 'Haru_firm']
for row in rows:
    rate = (values_20[row]+values_30[row]+values_40[row])/values_all[row]
    print(f"{row}'s rate : {rate}")
```

```python
# 20대의 검색량
search_20 = data_20[['Four_cuts', 'Photoism', 'Haru_firm']]

# 30대의 검색량
search_30 = data_30[['Four_cuts', 'Photoism', 'Haru_firm']]

# 40대의 검색량
search_40 = data_40[['Four_cuts', 'Photoism', 'Haru_firm']]

# 각 검색어 열에 대한 상대적인 검색량의 총합 계산
total_search_20 = search_20.sum(axis=0)
total_search_30 = search_30.sum(axis=0)
total_search_40 = search_40.sum(axis=0)

# 전체 검색량 총합 계산
total_all = total_search_20 + total_search_30 + total_search_40

# 각 검색어 열에 대한 전체 검색량에 대한 비율 계산
percentage_20 = total_search_20 / total_all * 100
percentage_30 = total_search_30 / total_all * 100
percentage_40 = total_search_40 / total_all * 100

# 결과 출력
print("20대의 검색어에 대한 전체 검색량에 대한 비율:")
print(percentage_20)
print("\n30대의 검색어에 대한 전체 검색량에 대한 비율:")
print(percentage_30)
print("\n40대의 검색어에 대한 전체 검색량에 대한 비율:")
print(percentage_40)
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import warnings

warnings.filterwarnings('ignore')

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/gulim.ttc"  # 한글 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 방문 비율이 1 미만인 데이터 필터링
filtered_data = foreigner_rank[foreigner_rank['방문비율(%)'] >= 1]
#print(sum(filtered_data['방문비율(%)']))
other_index = filtered_data[filtered_data['국가'] == '기타'].index

# '방문비율(%)' 값을 8.7로 변경
filtered_data.loc[other_index, '방문비율(%)'] = 8.7

# 필터링된 데이터
countries = filtered_data['국가']
visit_rates = filtered_data['방문비율(%)']

# 파이 차트 그리기
plt.figure(figsize=(10, 8))
plt.pie(visit_rates, labels=countries, autopct='%1.1f%%', startangle=0)
plt.title('Foreign Visitors by Country')
plt.axis('equal')  # 원형 차트를 그리기 위해 가로세로 비율을 같게 설정
plt.show()
```
